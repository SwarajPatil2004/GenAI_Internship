```python
import numpy as np
inputs= np.array([17, -18, -2, 9, -10])
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

def relu(x):
    return np.maximum(0,x)

def leaky_relu(x, alpha = 0.01):
    return np.where(x>0, x, alpha*x)

def softmax(x):
    e_x= np.exp(x - np.max(x))
    return e_x / e_x.sum(axis=0)
```

```python
activation_function= {"Sigmoid" : sigmoid(inputs),
                      "Tanh" : tanh(inputs),
                      "Relu" : relu(inputs),
                      "Leaky Relu" : leaky_relu(inputs),
                      "Softmax" : softmax(inputs)}
```

```python
for name, output in activation_function.items():
    print(name, ":\n", output, "\n")
```
